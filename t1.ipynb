```python
#Importing Library
from PIL import Image
from sys import argv
```

```python
source = "test.txt" # path of your text file
```

```python
# try:
#     txt=open(argv[1], "r")
# except IndexError:
#     print("No file entered. Using default file...")
#     txt=open(fileName, "r")
# except FileNotFoundError:
#     print("Could not find file. Using default file...")
#     txt=open(fileName, "r")
txt = open(source, "r")
```

```python
BG=Image.open("res/F_Pranav/bg.png") 
sheet_width=BG.width
gap, ht = 0, 0
```

```python
for i in txt.read().replace("\n",""):
        cases = Image.open("res/F_Pranav/{}.png".format(str(ord(i))))
        BG.paste(cases, (gap, ht))
        size = cases.width
        height=cases.height
        #print(size)
        print("Running...........")
        gap+=size
        

        if sheet_width < gap or len(i)*115 >(sheet_width-gap):
            gap,ht=0,ht+140

print(gap)
print(sheet_width)
BG.show()
```
